<hr style="border:1px solid #ddd;">

<h3>üìå Resposta</h3>

<p><b>Aeroelasticidade est√°tica</b> estuda o equil√≠brio entre:</p>

<ul>
<li><b>For√ßas aerodin√¢micas</b> (dependem do escoamento e da velocidade);</li>
<li><b>Deforma√ß√µes estruturais</b> (flex√£o e tor√ß√£o);</li>
<li><b>Rigidez da estrutura</b>.</li>
</ul>

<p>
Ela √© crucial porque a deforma√ß√£o estrutural altera o <b>√¢ngulo efetivo de ataque</b>, 
modificando as for√ßas aerodin√¢micas, que por sua vez provocam novas deforma√ß√µes ‚Äî 
podendo levar a <b>instabilidades</b>, como:
</p>

<ul>
<li><b>Diverg√™ncia</b> (crescimento sem limite do √¢ngulo de tor√ß√£o);</li>
<li><b>Revers√£o de comando</b> (perda de efici√™ncia ou invers√£o do efeito da superf√≠cie de controle).</li>
</ul>

</h2>

<hr style="border:1px solid #ddd;">

<h3>üìå Resposta</h3>

<h4>üîÅ An√°lise Iterativa</h4>

<p>
Na <b>an√°lise iterativa</b>, assume-se inicialmente um estado estrutural (por exemplo, um valor de tor√ß√£o da asa) e procede-se da seguinte forma:
</p>

<ol>
<li>Calculam-se as <b>for√ßas aerodin√¢micas</b>;</li>
<li>Determinam-se as <b>deforma√ß√µes estruturais</b>;</li>
<li>Atualiza-se a geometria (tor√ß√£o/flex√£o);</li>
<li>O processo √© repetido at√© <b>convergir</b> (regime est√°vel) ou <b>divergir</b> (instabilidade).</li>
</ol>

<p>
Esse m√©todo √© especialmente √∫til quando o modelo √© <b>mais complexo ou n√£o linear</b>, embora apresente maior custo computacional.
</p>

<h4>üìê An√°lise Direta</h4>

<p>
Na <b>an√°lise direta</b>, monta-se explicitamente a <b>equa√ß√£o aeroel√°stica acoplada</b>, obtendo-se a condi√ß√£o de diverg√™ncia de forma anal√≠tica ou semi-anal√≠tica.
</p>

<p>
A velocidade de diverg√™ncia √© identificada quando:
</p>

<ul>
<li>um <b>denominador da equa√ß√£o se anula</b>, ou</li>
<li>o <b>determinante do sistema √© igual a zero</b>.</li>
</ul>

<p>
Esse m√©todo √© <b>mais r√°pido</b> e fornece diretamente a <b>velocidade cr√≠tica de diverg√™ncia</b>, sendo muito utilizado em an√°lises preliminares de projeto.

<h2 style="color:#c62828;">
Qual a influ√™ncia do √¢ngulo de incid√™ncia inicial Œ∏<sub>0</sub> na velocidade de diverg√™ncia?
</h2>

<p>
Em <b>modelos lineares cl√°ssicos</b>, a <b>velocidade de diverg√™ncia</b> √© determinada pela
<b>rigidez estrutural</b> e pela <b>realimenta√ß√£o aeroel√°stica</b>
(o termo que multiplica <i>Œ∏</i>).
</p>

<p>
O termo <b>Œ∏<sub>0</sub></b> normalmente:
</p>

<ul>
<li>
<b>altera o valor da tor√ß√£o</b> para uma dada velocidade
(<i>amplitude da resposta</i>);
</li>
<li>
<b>n√£o altera</b> ‚Äî ou altera muito pouco ‚Äî a <b>velocidade cr√≠tica de diverg√™ncia</b>,
pois a diverg√™ncia decorre da <b>perda de rigidez efetiva do sistema</b>,
e n√£o da exist√™ncia de um <i>offset</i> inicial.
</li>
</ul>


<h2 style="color:#c62828;">
O que √© rigidez aeroel√°stica e qual a sua rela√ß√£o com a velocidade de diverg√™ncia?
</h2>

<hr style="border:1px solid #ddd;">

<h3>üìå Resposta</h3>

<p>
A <b>rigidez aeroel√°stica</b> representa a rigidez <b>efetiva</b> de uma estrutura quando se considera,
al√©m da rigidez puramente estrutural, os efeitos das <b>for√ßas aerodin√¢micas acopladas √† deforma√ß√£o</b>.
</p>

<p>
A estrutura possui uma rigidez ‚Äúpura‚Äù, por exemplo:
</p>

<ul>
<li><b>GJ</b> no caso de tor√ß√£o;</li>
<li><b>EI</b> no caso de flex√£o.</li>
</ul>

<p>
Entretanto, a <b>aerodin√¢mica introduz uma rigidez adicional</b>, que pode atuar como
<b>positiva ou negativa</b>, dependendo da posi√ß√£o do <b>centro aerodin√¢mico</b> em rela√ß√£o ao
<b>eixo el√°stico</b>.
</p>

<p>
Assim, a rigidez aeroel√°stica efetiva pode ser expressa como:
</p>

<p style="text-align:center;">
<b>K<sub>efetivo</sub> = K<sub>estrutural</sub> ‚àí K<sub>aero</sub>(V)</b>
</p>

<p>
√Ä medida que a velocidade aumenta, o termo aerodin√¢mico
<b>K<sub>aero</sub>(V)</b> cresce, reduzindo a rigidez efetiva do sistema.
</p>

<p>
Quando:
</p>

<p style="text-align:center;">
<b>K<sub>efetivo</sub> ‚Üí 0</b>
</p>

<p>
o sistema deixa de resistir ao carregamento aerodin√¢mico,
resultando em uma <b>instabilidade est√°tica</b>, conhecida como
<b>diverg√™ncia</b>.
</p>

<p>
Portanto, a <b>velocidade de diverg√™ncia</b> corresponde √† velocidade para a qual a
<b>rigidez aeroel√°stica efetiva se anula</b>.
</p>

<h2 style="color:#c62828;">
Atividade ‚Äî Aeroelasticidade Est√°tica (modelo simples)  
</h2>

<hr style="border:1px solid #ddd;">

<h3>üìå Par√¢metros do enunciado</h3>

<p>
Usar os par√¢metros: <b>C, S, œÅ, a<sub>w</sub>, Œ∏<sub>0</sub>, V, e, GJ</b> e gerar gr√°ficos do √¢ngulo de tor√ß√£o na ponta
<b>Œ∏<sub>tip</sub></b> variando <b>Œ∏<sub>0</sub></b>, <b>V</b>, <b>e</b> e <b>GJ</b>.
</p>

<hr style="border:1px solid #ddd;">

<h3>üß© Modelo simples (asa em balan√ßo, 1-DOF)</h3>

<p><b>Press√£o din√¢mica:</b></p>
<p style="text-align:center;">
<b>q = ¬Ω œÅ V¬≤</b>
</p>

<p><b>Sustenta√ß√£o 2D por unidade de envergadura:</b></p>
<p style="text-align:center;">
<b>L‚Ä≤ = q ¬∑ C ¬∑ a<sub>w</sub> (Œ∏<sub>0</sub> + Œ∏)</b>
</p>

<p><b>Torque aerodin√¢mico distribu√≠do (por unidade de envergadura) no eixo el√°stico:</b></p>
<p style="text-align:center;">
<b>t = L‚Ä≤ (eC) = q ¬∑ C¬≤ ¬∑ a<sub>w</sub> ¬∑ e ¬∑ (Œ∏<sub>0</sub> + Œ∏)</b>
</p>

<p>
Para uma viga engastada com torque distribu√≠do constante, a tor√ß√£o na ponta √©:
</p>
<p style="text-align:center;">
<b>Œ∏<sub>tip</sub> = (t ¬∑ S¬≤) / (2 GJ)</b>
</p>

<p><b>Substituindo e isolando Œ∏<sub>tip</sub>:</b></p>

<p style="text-align:center;">
<b>Œ∏<sub>tip</sub> = (K ¬∑ Œ∏<sub>0</sub>) / (1 ‚àí K)</b>
</p>

<p style="text-align:center;">
onde &nbsp; <b>K = (q ¬∑ C¬≤ ¬∑ a<sub>w</sub> ¬∑ e ¬∑ S¬≤) / (2 GJ)</b>
</p>

<hr style="border:1px solid #ddd;">

<h3>‚ö†Ô∏è Condi√ß√£o de diverg√™ncia</h3>

<p>
A diverg√™ncia ocorre quando o sistema perde rigidez efetiva:
</p>

<p style="text-align:center;">
<b>1 ‚àí K = 0  &nbsp; ‚áí  &nbsp; K = 1</b>
</p>

<ul>
<li><b>Se e = 0</b>, ent√£o <b>K = 0</b> ‚Üí n√£o h√° acoplamento aeroel√°stico ‚Üí <b>n√£o h√° diverg√™ncia</b>.</li>
<li><b>Se e &lt; 0</b> (interpreta√ß√£o comum: efeito estabilizante), ent√£o <b>K &lt; 0</b> ‚Üí <b>n√£o h√° diverg√™ncia</b>.</li>
</ul>


```python
import numpy as np
import matplotlib.pyplot as plt

# Par√¢metros base (do enunciado)
C = 1.0          # m
S = 10.0         # m (semi-envergadura)
rho = 1.225      # kg/m^3
aw = 2*np.pi     # 1/rad
theta0_base = 0.1 # rad
v_base = 100.0    # m/s
e_base = 0.2      # adimensional (fra√ß√£o da corda)
GJ_base = 1e6     # N*m^2 (unidade f√≠sica t√≠pica para rigidez torsional)

def theta_tip(theta0, v, e, GJ):
    """Calcula tor√ß√£o na ponta usando o modelo 1-DOF.
       Retorna np.nan quando entra em diverg√™ncia (1-K <= 0 para e>0)."""
    q = 0.5 * rho * v**2
    K = (q * C**2 * aw * e * S**2) / (2.0 * GJ)
    denom = (1.0 - K)
    # Crit√©rio simples: se denom <= 0 e e>0 => diverg√™ncia (modelo prev√™ infinito)
    if np.isscalar(denom):
        if (e > 0) and (denom <= 0):
            return np.nan
        return (K * theta0) / denom
    else:
        out = (K * theta0) / denom
        out[(e > 0) & (denom <= 0)] = np.nan
        return out

def v_divergencia(e, GJ):
    """Velocidade de diverg√™ncia (quando K=1), se existir.
       Se e<=0 -> n√£o h√° diverg√™ncia (retorna np.inf)."""
    if e <= 0:
        return np.inf
    # K = (0.5*rho*V^2 * C^2 * aw * e * S^2)/(2 GJ) = 1
    # => V^2 = (4 GJ)/(rho*C^2*aw*e*S^2)
    V2 = (4.0 * GJ) / (rho * C**2 * aw * e * S**2)
    return np.sqrt(V2)

# 1) theta0 variando (ajustei para algo f√≠sico)
theta0_vec = np.arange(-0.5, 0.5001, 0.01)
theta_tip_theta0 = theta_tip(theta0_vec, v_base, e_base, GJ_base)

plt.figure()
plt.plot(theta0_vec, theta_tip_theta0)
plt.xlabel("theta0 [rad]")
plt.ylabel("theta_tip [rad]")
plt.title("Tor√ß√£o na ponta vs theta0 (v=100 m/s, e=0.2, GJ=1e6)")
plt.grid(True)

# 2) v variando
v_vec = np.arange(0.0, 500.0+1.0, 1.0)
theta_tip_v = theta_tip(theta0_base, v_vec, e_base, GJ_base)
Vdiv = v_divergencia(e_base, GJ_base)

plt.figure()
plt.plot(v_vec, theta_tip_v)
if np.isfinite(Vdiv) and Vdiv <= v_vec.max():
    plt.axvline(Vdiv, linestyle="--", label=f"Vdiv ‚âà {Vdiv:.1f} m/s")
    plt.legend()
plt.xlabel("v [m/s]")
plt.ylabel("theta_tip [rad]")
plt.title("Tor√ß√£o na ponta vs velocidade")
plt.grid(True)

# 3) e variando
e_vec = np.arange(-0.3, 0.3001, 0.01)
theta_tip_e = np.array([theta_tip(theta0_base, v_base, ee, GJ_base) for ee in e_vec])

plt.figure()
plt.plot(e_vec, theta_tip_e)
plt.axvline(0.0, linestyle="--", label="e=0 (CE=CA)")
plt.legend()
plt.xlabel("e [-] (offset CE-CA em fra√ß√£o de C)")
plt.ylabel("theta_tip [rad]")
plt.title("Tor√ß√£o na ponta vs e (v=100 m/s)")
plt.grid(True)

# 4) GJ variando (logspace √© melhor para 1 a 1e8)
GJ_vec = np.logspace(0, 8, 200)  # 1 a 1e8
theta_tip_GJ = theta_tip(theta0_base, v_base, e_base, GJ_vec)

plt.figure()
plt.semilogx(GJ_vec, theta_tip_GJ)
plt.xlabel("GJ [N*m^2] (escala log)")
plt.ylabel("theta_tip [rad]")
plt.title("Tor√ß√£o na ponta vs GJ (v=100 m/s, e=0.2)")
plt.grid(True)

plt.show()

# Demonstra√ß√µes pedidas:
# - Quanto menor |e| (aproxima CE de CA) ou maior GJ => maior Vdiv
for ee in [0.05, 0.1, 0.2, 0.3]:
    print(f"e={ee:.2f} -> Vdiv={v_divergencia(ee, GJ_base):.2f} m/s")

for GJ in [1e5, 1e6, 1e7]:
    print(f"GJ={GJ:.1e} -> Vdiv={v_divergencia(e_base, GJ):.2f} m/s")

# - Se CE=CA (e=0) n√£o h√° diverg√™ncia:
print("e=0 -> Vdiv =", v_divergencia(0.0, GJ_base))

# - CE > CA (aqui representado como e<0) n√£o h√° diverg√™ncia:
print("e=-0.1 -> Vdiv =", v_divergencia(-0.1, GJ_base))
```

<h3>1) ‚ÄúQuanto menor a dist√¢ncia entre CE e CA ou maior GJ, maior a velocidade de diverg√™ncia‚Äù</h3>

<p>
No modelo implementado no c√≥digo, a diverg√™ncia ocorre quando <b>1 ‚àí K = 0</b>, isto √©, quando <b>K = 1</b>, com:
</p>

<p style="text-align:center;">
<b>K = (q ¬∑ C¬≤ ¬∑ a<sub>w</sub> ¬∑ e ¬∑ S¬≤) / (2GJ)</b> &nbsp; e &nbsp; <b>q = ¬Ω œÅ V¬≤</b>
</p>

<p>
Isolando <b>V</b> da condi√ß√£o <b>K = 1</b>, obt√©m-se a velocidade de diverg√™ncia:
</p>

<p style="text-align:center;">
<b>V<sub>div</sub> = ‚àö( 4GJ / (œÅ ¬∑ C¬≤ ¬∑ a<sub>w</sub> ¬∑ e ¬∑ S¬≤) )</b>
</p>

<p>
Portanto:
</p>

<ul>
<li>
<b>Diminuir e</b> (com <b>e &gt; 0</b>) <b>aumenta V<sub>div</sub></b>
(<i>pois e est√° no denominador</i>);
</li>
<li>
<b>Aumentar GJ</b> <b>aumenta V<sub>div</sub></b>
(<i>pois GJ est√° no numerador</i>).
</li>
</ul>

<p>
No c√≥digo, isso aparece diretamente na fun√ß√£o <b>V_divergencia(e, GJ)</b> e tamb√©m √© visualizado no gr√°fico
<b>Œ∏<sub>tip</sub> vs V</b>: ao reduzir <b>e</b> (com e&gt;0) ou aumentar <b>GJ</b>, o crescimento abrupto de Œ∏<sub>tip</sub>
(se aproximando da diverg√™ncia) ocorre em velocidades maiores.
</p>

<hr style="border:1px solid #ddd;">

<h3>2) ‚ÄúSe CE = CA n√£o h√° diverg√™ncia‚Äù</h3>

<p>
Quando <b>CE = CA</b>, interpreta-se <b>e = 0</b>. Logo:
</p>

<p style="text-align:center;">
<b>e = 0 ‚áí K = 0</b>
</p>

<p>
Com <b>K = 0</b>, n√£o existe condi√ß√£o em que <b>K = 1</b>. Assim, o denominador <b>1 ‚àí K</b> nunca zera e
<b>n√£o h√° diverg√™ncia</b> no modelo.
</p>

<hr style="border:1px solid #ddd;">

<h3>3) ‚ÄúCE &gt; CA n√£o h√° diverg√™ncia‚Äù</h3>

<p>
Interpretando <b>CE &gt; CA</b> como <b>e &lt; 0</b>, ent√£o:
</p>

<p style="text-align:center;">
<b>e &lt; 0 ‚áí K &lt; 0</b>
</p>

<p>
Se <b>K &lt; 0</b>, ent√£o <b>1 ‚àí K</b> √© sempre maior que 1 e <b>n√£o pode zerar</b>. Portanto, a condi√ß√£o de diverg√™ncia
<b>(1 ‚àí K = 0)</b> nunca √© atingida, e o sistema permanece est√°vel no contexto do modelo linear.
</p>

Quest√£o ‚Äî Implementa√ß√£o num√©rica do Exemplo (Item 8.3, 1¬™ edi√ß√£o, p. 129)

<p>
<b>Refa√ßa o exemplo do item 8.3</b> da primeira edi√ß√£o do livro texto (p√°gina 129) e

<p>
<b>Adote valores razo√°veis</b> para as vari√°veis do problema e determine:
</p>

<ul>
<li>o <b>√¢ngulo de tor√ß√£o na ponta da asa</b> (<b>Œ∏<sub>tip</sub></b>);</li>
<li>o <b>√¢ngulo de incid√™ncia efetivo</b> (<b>Œ±</b>);</li>
</ul>

<p>
para <b>diferentes valores da velocidade da aeronave</b>.

<p style="font-size:0.95em; color:#555;">
<b>Entrega esperada:</b> apresenta√ß√£o do modelo adotado, implementa√ß√£o num√©rica (c√≥digo),
e gr√°ficos de <b>Œ∏<sub>tip</sub>(V)</b> e <b>Œ±(V)</b> para a faixa de velocidades escolhida.
</p>


```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# TOR√á√ÉO E DIVERG√äNCIA ‚Äì EFEITO DA TRIMAGEM
# =========================

# Par√¢metros
C = 1.0          # m
S = 10.0         # m
rho = 1.225      # kg/m^3
aw = 2*np.pi
e = 0.2
GJ = 1e6         # N¬∑m^2/rad
W = 300000.0     # N

# Velocidade
v = np.arange(0.0, 501.0, 1.0)   # 0:1:500 (MATLAB)

# Press√£o din√¢mica
q = 0.5 * rho * v**2

# Press√µes cr√≠ticas
q_w = (3 * GJ) / (e * C**2 * S**2 * aw)
v_w = np.sqrt(2 * q_w / rho)

q_aer = 4 * q_w
v_aer = np.sqrt(2 * q_aer / rho)

# =========================
# Incid√™ncia e tor√ß√£o
# =========================
# Para evitar divis√£o por zero em v=0, usamos NaN
theta_0 = np.full_like(v, np.nan, dtype=float)
theta_T = np.full_like(v, np.nan, dtype=float)

mask = v > 0  # evita q=0

theta_0[mask] = (
    W * (1 - q[mask] / q_w)
    / (2 * q[mask] * C * S * aw * (1 - q[mask] / (4 * q_w)))
)

theta_T[mask] = (
    W * e * C * S
    / (4 * GJ * (1 - q[mask] / (4 * q_w)))
)

# =========================
# Gr√°ficos
# =========================
plt.figure(figsize=(10, 4))

# Incid√™ncia
plt.subplot(1, 2, 1)
plt.plot(v, theta_0)
plt.title('Incid√™ncia em fun√ß√£o da velocidade')
plt.xlabel('V [m/s]')
plt.ylabel(r'$\theta_0$ [rad]')
plt.grid(True)

# Tor√ß√£o na ponta
plt.subplot(1, 2, 2)
plt.plot(v, theta_T)
plt.title('Tor√ß√£o na ponta de asa em fun√ß√£o da velocidade')
plt.xlabel('V [m/s]')
plt.ylabel(r'$\theta_T$ [rad]')
plt.grid(True)

plt.tight_layout()
plt.show()

# =========================
# Impress√£o de velocidades cr√≠ticas (opcional)
# =========================
print(f"Velocidade de diverg√™ncia por tor√ß√£o (v_w) ‚âà {v_w:.2f} m/s")
print(f"Velocidade aerodin√¢mica cr√≠tica (v_aer) ‚âà {v_aer:.2f} m/s")
```

<h3>4.1 Efeito do enflechamento na aerodin√¢mica</h3>
